```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
gc = pd.read_csv('gen_cap_merged.csv')
gc.set_index('date', inplace=True)
gc.index = pd.to_datetime(gc.index)

cons = pd.read_csv('cons.csv')
cons.set_index('date', inplace=True)
```

```python

fig, ax = plt.subplots(figsize=(10, 8))

ax.plot(gc.index, gc['total_gen'], c='green', label='Electricity Generation')
ax.plot(gc.index, cons['total_cons'], c='red', label='Electricity Demand')
ax.plot(gc.index, cons['moving_avg'], c='black', label='Demand Moving Average', linestyle='--')
ax.plot(gc.index, gc['tot_moving_avg'], c='blue', label='Generation Moving Average', linestyle='--')
plt.title("Electricity Generation v. Demand, Taiwan, 2003-2023")
plt.xlabel("Date")
plt.ylabel("kWh")
plt.legend(loc='upper left')

myFmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_formatter(myFmt)
plt.show()
```
